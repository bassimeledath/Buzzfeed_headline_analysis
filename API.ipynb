```python
import requests
import pandas as pd
import numpy as np
```

```python
response = requests.get("https://newsapi.org/v2/everything?sources=buzzfeed&apiKey=a9664bd87f964aa8a68d9fe6285d708e")
```

```python
first = [x for x in str(response.content).split('"') if len(x)>20]
second = [x for x in first if ('http' not in x) & ('chars' not in x)]
third = [x.replace("View Entire Post","") for x in second]
fourth = [x.replace("\\xe2\\x80\\xba","") for x in third]
fifth = [x.replace("\\","") for x in fourth]
sixth = [x.replace("xe2x80x98","") for x in fifth]
seventh = [x.replace("xe2x80x99","") for x in sixth]
final = pd.DataFrame([x for x in seventh if len(x)>2])
```
